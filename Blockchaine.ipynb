```python
import datetime # Permet de travailler avec les dates et heures
import hashlib  # Permet de créer des 'empreintes' uniques (hash) pour les blocs
```

# Class blockchain


```python
class Block: # Définit la classe Block (un bloc de la chaîne)
  def __init__(self, index, data, previousHash, difficulty, timestamp=None):
    self.index = index # Numéro unique du bloc
    self.data = data # Données stockées dans ce bloc
    self.previousHash = previousHash # Hash (empreinte) du bloc précédent
    self.difficulty = difficulty # Difficulté du minage (nombre de zéros au début du hash)
    self.timestamp = timestamp if timestamp else datetime.datetime.now() # Date et heure de création du bloc
    self.nonce = 0 # Nombre utilisé une seule fois pour trouver le hash valide
    self.hash = self.mineBlock() # Le hash (empreinte) de ce bloc, calculé par le minage

  def calculateHash(self): # Calcule l'empreinte (hash) du bloc
    block_string = (
      str(self.index)
      + str(self.timestamp)
      + str(self.data)
      + str(self.previousHash)
      + str(self.nonce)
  )
    return hashlib.sha256(block_string.encode("utf-8")).hexdigest() # Retourne le hash en utilisant SHA256

  def mineBlock(self): # Trouve le hash du bloc en résolvant une énigme (preuve de travail)
    target = "0" * self.difficulty # Le hash doit commencer par un certain nombre de zéros
    while True: # Boucle infinie jusqu'à trouver le bon hash
      candidate = self.calculateHash() # Calcule un hash possible
      if candidate.startswith(target): # Si le hash commence par le bon nombre de zéros
        return candidate # On a trouvé le bon hash, on le retourne
      self.nonce += 1 # Sinon, on change le 'nonce' et on essaie à nouveau

  def __str__(self): # Représentation textuelle du bloc, utile pour l'affichage
    return (
      f"Block(index={self.index}, timestamp={self.timestamp}, "
      f"data={self.data}, previousHash={self.previousHash}, "
      f"nonce={self.nonce}, difficulty={self.difficulty}, hash={self.hash})"
  )
```

Class blockchain


```python
class Blockchain: # Définit la classe Blockchain (la chaîne de blocs)
  def __init__(self, difficulty=3):
    self.index = 0 # Index du dernier bloc ajouté
    self.difficulty = difficulty # Difficulté de minage pour tous les blocs de cette chaîne
    self.chain = [
    Block(0, "blalalal", "0" * 64, self.difficulty, timestamp=datetime.datetime.now()) # Le premier bloc, appelé "bloc de genèse"
]

  def getLastBlock(self): # Renvoie le dernier bloc de la chaîne
    return self.chain[-1]

  def addBlock(self, data): # Ajoute un nouveau bloc à la chaîne
    self.index += 1 # Incrémente l'index du bloc
    new_block = Block( # Crée un nouveau bloc
      index=self.index, # Son numéro d'index
      data=data, # Les données à y stocker
      previousHash=self.getLastBlock().hash, # L'empreinte du bloc précédent
      difficulty=self.difficulty # La difficulté de minage
  )
    self.chain.append(new_block) # Ajoute le nouveau bloc à la chaîne

  def checkChainValid(self): # Vérifie si toute la chaîne est valide
    for i in range(1, len(self.chain)): # Parcourt tous les blocs à partir du deuxième
      current = self.chain[i] # Le bloc actuel
      previous = self.chain[i - 1] # Le bloc précédent
      if current.calculateHash() != current.hash: # Vérifie si l'empreinte du bloc actuel est correcte
        return False # Si non, la chaîne est invalide
      if current.previousHash != previous.hash: # Vérifie si le lien avec le bloc précédent est correct
        return False # Si non, la chaîne est invalide
    return True # Si toutes les vérifications passent, la chaîne est valide

  def displayChain(self): # Affiche tous les blocs de la chaîne
    for block in self.chain: # Pour chaque bloc dans la chaîne
      print(block) # Affiche le bloc
      print("-" * 90) # Affiche une ligne de séparation
```

```python
my_blockchain = Blockchain() # Crée une nouvelle blockchain
print("Mining block 1...") # Affiche un message
my_blockchain.addBlock("data block 1") # Ajoute le premier bloc de données
print("Mining block 2...") # Affiche un message
my_blockchain.addBlock("data block 2") # Ajoute le deuxième bloc de données
print("Mining block 3...") # Affiche un message
my_blockchain.addBlock("data block 3") # Ajoute le troisième bloc de données

print("\nBlockchain:") # Affiche un titre
my_blockchain.displayChain() # Affiche tous les blocs de la blockchain

print(f"\nIs blockchain valid? {my_blockchain.checkChainValid()}") # Vérifie et affiche si la blockchain est valide
```
